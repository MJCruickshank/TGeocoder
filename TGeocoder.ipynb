```python
from utils import get_recent_messages, build_map, parse_results, geocode_results_openai
from telethon import TelegramClient
from openai import OpenAI
```

```python
API_ID = '' #Your Telegram API ID (str)
API_HASH = '' #Your Telegram API Hash (str)
PHONE_NUMBER = '' #Your Telegram-associated phone number (str)
CHANNEL_USERNAME = '' #The username of the Telegram channel you wish to parse
NUM_MESSAGES = 40 #The number of messages you wish to parse from the channel (int)

#Your OpenAI API Key
OPENAI_API_KEY = ''
```

```python
# Initalize the OpenAI client
openai_client = OpenAI(api_key = OPENAI_API_KEY)

# Initialize the Telegram client
tg_client = TelegramClient('your_session_name', API_ID, API_HASH)
tg_client.start(phone=PHONE_NUMBER)
```

```python
messages = await get_recent_messages(tg_client, CHANNEL_USERNAME, NUM_MESSAGES)
```

```python
messages = [x for x in messages if len(x.message)>0] #Eliminate blank messages
message_texts = [x.message for x in messages]
message_dates = [x.date for x in messages]
```

```python
parsed_results_df = parse_results(openai_client, message_texts, message_dates)
geolocated_events = geocode_results_openai(parsed_results_df, openai_client)
```
